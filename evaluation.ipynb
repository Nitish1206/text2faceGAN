```python
import pandas as pd
from torchvision.utils import save_image
import shutil
import os
from glob import glob
from tqdm import tqdm
```

```python
from image_generator import ImageGenerator
```

```python
gen_model=ImageGenerator()
```

```python
csv_file_path="evaluation/text_5_descr_celeba.csv"
```

```python
df=pd.read_csv(csv_file_path)
```

```python
df.head()
```

```python
df.shape
```

```python
sample_size=15000
```

```python
sample_df=df.sample(sample_size)
```

```python
sample_df.head()
```

```python
image_path= "evaluation\CelebA\img_align_celeba\img_align_celeba"
```

```python
gen_image_path="evaluation/gen/"
original_eval_image="evaluation/eval_image/"
```

```python
for gen_pth,eval_pth in zip(glob(gen_image_path+"/*"),glob(original_eval_image+"/*")):
    os.remove(gen_pth)
    os.remove(eval_pth)
```

```python
for image_name,text_desc in tqdm(zip(sample_df["image_id"],sample_df["text_description"])):
    frame=gen_model.gen(text_desc)
    save_image(frame, gen_image_path+str(image_name)+".png")
    shutil.copy(image_path+os.sep+image_name, original_eval_image+image_name)
    
    
```
